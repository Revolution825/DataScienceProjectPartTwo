Brief Description:
This notebook aims to use a Linear SVM to try and predict if a particular email example is a phishing email or not.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
dataset = pd.read_csv("PhiUSIIL_Phishing_URL_Dataset.csv")
dataset.head()
```

```python
dataset = dataset.drop(columns=["FILENAME", "URL", "Domain", "TLD", "Title"])
```

```python
# Shows correalation coefficients for predicting the label
correlation_matrix = dataset.corr()
correlation_matrix["label"]
```

```python
X = dataset[["URLSimilarityIndex", "HasSocialNet"]]
X = X.values

y = dataset["label"]
y = y.values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=12345
)
```

```python
# Scaling data so everything is in a comparable range
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svm_model = LinearSVC(random_state=12345)
```

```python
svm_model.fit(X_train_scaled, y_train)
```

```python
y_pred = svm_model.predict(X_test_scaled)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print(classification_report(y_test, y_pred))
```
